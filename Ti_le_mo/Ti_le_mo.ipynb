```python
cate_of_TLM = []# Categories of BMI
for i in range(len(file['Tỉ lệ mỡ cơ thể'])):
  if file['Giới tính'][i] == 1:
    if file['Tỉ lệ mỡ cơ thể'][i] >= 26:
      cate_of_TLM.append(0)#'Thừa cân béo phì'
    elif 18 <= file['Tỉ lệ mỡ cơ thể'][i] <= 25:
      cate_of_TLM.append(1)#'Bình thường'
    elif 14<= file['Tỉ lệ mỡ cơ thể'][i] <= 17:
      cate_of_TLM.append(2)#'fit'
    elif 6<= file['Tỉ lệ mỡ cơ thể'][i]  <= 13:
      cate_of_TLM.append(3)#'vận động viên'
    else:
      cate_of_TLM.append(4)#'Lượng mỡ thiết yếu'
  else:
    if file['Tỉ lệ mỡ cơ thể'][i] >= 32:
      cate_of_TLM.append(0)#'Thừa cân béo phì'
    elif 25 <= file['Tỉ lệ mỡ cơ thể'][i] <= 31:
      cate_of_TLM.append(1)#'Bình thường'
    elif 21<= file['Tỉ lệ mỡ cơ thể'][i] <= 24:
      cate_of_TLM.append(2)#'fit'
    elif 14<= file['Tỉ lệ mỡ cơ thể'][i]  <= 20:
      cate_of_TLM.append(3)#'Vận động viên'
    else:
      cate_of_TLM.append(4)#'Lượng mỡ thiết yếu'

file['Categories of TLM'] = cate_of_TLM
```

```python
Y = file['Categories of TLM'].values
X = file.drop( redisual, axis  = 1)
print(Y.shape)
print(X.shape)
```

```python
X = X.astype(float)
mycols = X.columns
```

Standardized data : min-max scaling, StandardScaler

1.   List item
2.   List item



```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
std = StandardScaler()
mms = MinMaxScaler()
X_Scaler = std.fit_transform(X)
X = pd.DataFrame(X_Scaler)
#X.columns = mycols
print(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25)
```

```python
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
```

#1. Random forest


```python
RFC = RandomForestClassifier()
RFC.fit(X_train, Y_train)
print(RFC.score(X_test, Y_test))
```

```python
print(classification_report(Y_test,RFC.predict(X_test),labels=[0,1,2,3,4],digits = 4))
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
```

```python
print("training set")
print(mean_squared_error(Y_train, RFC.predict(X_train), squared=False))
print(accuracy_score(Y_train, RFC.predict(X_train)))


print("testing set")
print(mean_squared_error(Y_test, RFC.predict(X_test), squared=False))
print(accuracy_score(Y_test, RFC.predict(X_test)))
```

#5. ANN

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(50, input_dim = len(X_train.values[0]), activation= 'relu'))
model.add(Dense(25, activation='sigmoid'))
model.add(Dense(25, activation='sigmoid'))
model.add(Dense(1, activation='linear'))
```

```python
model.compile(loss= 'mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs=50, batch_size=5)
```

```python
raw_output = model.predict(X_test)

predict = []
for y in raw_output:
    if y[0] >= 4:
      predict.append(4)
    else:
      predict.append(round(y[0]))
```

```python
print(classification_report(Y_test, predict, labels=[0,1,2,3,4],digits = 4))
```
